```python
# If you're in Google Colab, uncomment these:
# !pip -q install ipywidgets

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from ipywidgets import interact, widgets
from IPython.display import display

sns.set_theme(style="whitegrid")
```

```python
df = sns.load_dataset("tips")
df.head()
```

Data Preprocessing

Check missing values

Validate data types

Create feature: tip percentage

```python
print("Shape:", df.shape)
print("\nMissing values:\n", df.isnull().sum())

df["tip_pct"] = (df["tip"] / df["total_bill"]) * 100

# Basic type check
print("\nDtypes:\n", df.dtypes)

df.describe()
```


Interactive Filters

```python
days = ["All"] + sorted(df["day"].unique().tolist())
times = ["All"] + sorted(df["time"].unique().tolist())
smokers = ["All"] + sorted(df["smoker"].unique().tolist())

day_dd = widgets.Dropdown(options=days, value="All", description="Day:")
time_dd = widgets.Dropdown(options=times, value="All", description="Time:")
smoker_dd = widgets.Dropdown(options=smokers, value="All", description="Smoker:")

bill_min, bill_max = float(df["total_bill"].min()), float(df["total_bill"].max())
bill_slider = widgets.FloatRangeSlider(
    value=[bill_min, bill_max],
    min=bill_min,
    max=bill_max,
    step=1.0,
    description="Bill $:",
    continuous_update=False,
    readout_format=".0f",
    layout=widgets.Layout(width="70%")
)

ui = widgets.VBox([widgets.HBox([day_dd, time_dd, smoker_dd]), bill_slider])
display(ui)

def filter_df(day, time, smoker, bill_range):
    dff = df.copy()
    if day != "All":
        dff = dff[dff["day"] == day]
    if time != "All":
        dff = dff[dff["time"] == time]
    if smoker != "All":
        dff = dff[dff["smoker"] == smoker]
    dff = dff[(dff["total_bill"] >= bill_range[0]) & (dff["total_bill"] <= bill_range[1])]
    return dff

def show_preview(day, time, smoker, bill_range):
    dff = filter_df(day, time, smoker, bill_range)
    display(dff.head(10))
    print("Filtered rows:", len(dff))

out = widgets.interactive_output(show_preview, {
    "day": day_dd, "time": time_dd, "smoker": smoker_dd, "bill_range": bill_slider
})
display(out)
```

Visualization 1: Bar Chart (Average Tip by Day)

```python
def plot_avg_tip(day, time, smoker, bill_range):
    dff = filter_df(day, time, smoker, bill_range)
    if len(dff) == 0:
        print("No data for selected filters.")
        return

    plt.figure(figsize=(8,5))
    sns.barplot(data=dff, x="day", y="tip", estimator=np.mean, ci=None)
    plt.title("Average Tip by Day (Filtered)")
    plt.xlabel("Day")
    plt.ylabel("Average Tip ($)")
    plt.show()

    # Interpretation
    grp = dff.groupby("day")["tip"].mean().sort_values(ascending=False)
    print("Interpretation:")
    print("- Highest average tip day:", grp.index[0], f"(${grp.iloc[0]:.2f})")
    print("- Lowest average tip day:", grp.index[-1], f"(${grp.iloc[-1]:.2f})")

widgets.interactive_output(plot_avg_tip, {
    "day": day_dd, "time": time_dd, "smoker": smoker_dd, "bill_range": bill_slider
})
```

Visualization 2: Scatter Plot (Total Bill vs Tip)

```python
color_by = widgets.Dropdown(
    options=["time", "day", "sex", "smoker", "size"],
    value="time",
    description="Color by:"
)
display(color_by)

def plot_scatter(day, time, smoker, bill_range, color_col):
    dff = filter_df(day, time, smoker, bill_range)
    if len(dff) == 0:
        print("No data for selected filters.")
        return

    plt.figure(figsize=(8,5))
    sns.scatterplot(data=dff, x="total_bill", y="tip", hue=color_col)
    plt.title(f"Total Bill vs Tip (Colored by {color_col})")
    plt.xlabel("Total Bill ($)")
    plt.ylabel("Tip ($)")
    plt.legend(title=color_col, bbox_to_anchor=(1.02, 1), loc="upper left")
    plt.show()

    # Interpretation (simple correlation)
    corr = dff["total_bill"].corr(dff["tip"])
    print("Interpretation:")
    print(f"- Correlation between total bill and tip: {corr:.2f} (higher means stronger positive relationship)")

widgets.interactive_output(plot_scatter, {
    "day": day_dd, "time": time_dd, "smoker": smoker_dd, "bill_range": bill_slider, "color_col": color_by
})
```

Visualization 3: Line Chart (Average Total Bill by Party Size)

```python
def plot_line(day, time, smoker, bill_range):
    dff = filter_df(day, time, smoker, bill_range)
    if len(dff) == 0:
        print("No data for selected filters.")
        return

    summary = dff.groupby("size", as_index=False)["total_bill"].mean().sort_values("size")

    plt.figure(figsize=(8,5))
    plt.plot(summary["size"], summary["total_bill"], marker="o")
    plt.title("Average Total Bill by Party Size (Filtered)")
    plt.xlabel("Party Size")
    plt.ylabel("Average Total Bill ($)")
    plt.xticks(summary["size"])
    plt.show()

    print("Interpretation:")
    if len(summary) >= 2:
        print(f"- Bills generally {'increase' if summary['total_bill'].iloc[-1] > summary['total_bill'].iloc[0] else 'do not increase'} with party size.")
    print("- This helps estimate staffing needs and expected revenue by table size.")

widgets.interactive_output(plot_line, {
    "day": day_dd, "time": time_dd, "smoker": smoker_dd, "bill_range": bill_slider
})
```

Optional Visualization 4: Distribution of Tip Percentage

```python
def plot_tip_pct_dist(day, time, smoker, bill_range):
    dff = filter_df(day, time, smoker, bill_range)
    if len(dff) == 0:
        print("No data for selected filters.")
        return

    plt.figure(figsize=(8,5))
    sns.histplot(dff["tip_pct"], kde=True)
    plt.title("Distribution of Tip Percentage (Filtered)")
    plt.xlabel("Tip Percentage (%)")
    plt.ylabel("Count")
    plt.show()

    print("Interpretation:")
    print(f"- Median tip %: {dff['tip_pct'].median():.1f}%")
    print(f"- 90th percentile tip %: {dff['tip_pct'].quantile(0.90):.1f}%")
    print("- Outliers may indicate exceptional service, discounts, or data noise.")

widgets.interactive_output(plot_tip_pct_dist, {
    "day": day_dd, "time": time_dd, "smoker": smoker_dd, "bill_range": bill_slider
})
```

Summary + Business Implications
Key Findings

Spending distribution is typically skewed (many small bills, few large).

Bill size and tip show a positive relationship.

Day/time/party size drive differences in customer behavior and revenue.
